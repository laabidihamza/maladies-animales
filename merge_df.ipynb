```python
import pandas as pd
```

```python
df_traduit = pd.read_csv('data/dataset_traduit.csv')
df_original = pd.read_csv('data/output_dataset.csv')
```

```python
# Merge only the 3 columns from source_df into target_df
merged_df = df_traduit.merge(
    df_original[['code', 'resume_50_mots', 'resume_100_mots', 'resume_150_mots']],
    on='code',
    how='left'  # Use 'left' to keep all rows from target_df
)
```

```python
merged_df.to_csv('data/new_dataset_with_resume_columns.csv', index=False)
```
